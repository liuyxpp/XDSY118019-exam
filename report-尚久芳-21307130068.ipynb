In mathematics, the <strong>Hénon map</strong> is a discrete-time dynamical system. It is one of the most studied examples of dynamical systems that exhibit <a href="https://en.wikipedia.org/wiki/Chaos_theory" title="Chaos theory">chaotic behavior</a>. the Hénon map is given by

$x_{n+1} = 1 - ax^2_n+y_n$ \
$y_{n+1} = bx_n$ \
This map produces an orbit iteratively, i.e. starting from $u_0=(x_0,y_0)$ one applies the above rule to get $u_1=(x_1,y_1)$, then applies the rule again on $u_1$ to get $u_2$, and so on. The orbit consists of the sequence of states $[u_0,u_1,u_2,...]$.

## 1. 计算 Henon maps 的函数
计算Hénon map的函数：输入任意函数系数<code>a, b</code>、初始值<code>u0</code>以及轨迹长度<code>N</code>，输出Hénon map的轨迹<code>[u0, u1, u2, ..., uN]</code>。

```python
def get(a, b, u, n):
    l =[u,]
    v = [0,0]
    for i in range(n):
        v[0] = 1 - a * u[0]**2 + u[1]
        v[1] = b * u[0]
        u = tuple(v)
        l.append(u)
    return l
```

## 2. 绘制轨迹图
利用编写的函数计算经典Hénon map的轨迹，参数取值为<code>a=1.4, b=0.3, u0=(0, 0)</code>，探索<code>N</code>的取值，求解得到的轨迹，并用绘制轨迹图（<code>x</code>为横坐标，<code>y</code>为纵坐标）。

```python
import matplotlib.pyplot as plt

a = 1.4
b = 0.3
u = (0,0)
n = 9911
l = get(a, b, u, n)

def draw(l):
    # 提取 x 和 y 坐标
    x = [point[0] for point in l]
    y = [point[1] for point in l]
    # 创建图形和轴
    plt.scatter(x, y, marker='o', color='blue', label="Henon Map")  # 绘制散点图
    plt.xlabel("X-axis")  # X轴标签
    plt.ylabel("Y-axis")  # Y轴标签
    plt.title("Henon Maps")  # 标题
    plt.legend()  # 图例
    plt.grid(True)  # 添加网格
    plt.show()  # 显示图形

draw(l)
```

## 3. 绘制 orbit digram 图
利用编写的函数计算Hénon map的orbit digram，即固定<code>b=0.3</code>，改变<code>a</code>后获得一系列Hénon map的轨迹，然后以<code>a</code>为横轴，<code>x</code>为纵轴绘制orbit digram图。

```python
import numpy as np

b = 0.3
last = 1000
epoch = 10
a = np.linspace(0.5, 1.5, epoch)
for i in range(epoch):
    l = get(a[i], b, u, n)
    for j in range(last):
        plt.plot(a[i], l[-j][0], 'o', color='red', markersize=1)
# 设置图表标题和标签
plt.title("Orbit Diagram of Henon Map")
plt.xlabel("Parameter a")
plt.ylabel("x values")
plt.show()
```

## 4. 找出 Henon maps 收敛的 a
分析上述画出的orbit digram，找到Hénon map可以收敛到一条周期性轨道的<code>a</code>值，计算该<code>a</code>值对应的Hénon map的轨迹并绘图。

```python
a = 1.2
l = get(a, b, u, n)
draw(l)
```
