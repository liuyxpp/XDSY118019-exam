```python
##任务1：输出pi值
import numpy as np
import matplotlib.pyplot as plt
m = 0
for i in range(0,N):
 a1 = np.random.random()
 a2 = np.random.random()
 if (a1*a1)+(a2*a2)<= 1:
    m=m+1
ans = 4*m/N
ans
```

```python
#任务2画出100
N=100
l1 = []
l2 = []
for i in range(0,N):
 rr = np.random.random(N)
 l1.append(rr)
for i in range(0,N):
 rr = np.random.random(N)
 l2.append(rr)
theta = np.linspace(0, 2 * np.pi, 200)
x = np.cos(theta)
y = np.sin(theta)
fig, ax = plt.subplots(figsize=(4, 4))
ax.plot(x, y, color="darkred", linewidth=2)
sizes = np.linspace(0.5, 1.0, num=N)
for i in range(0,N):
 plt.scatter(l1[i],l2[i],s = sizes)
plt.xlim((0,1.5))
plt.ylim((0,1.5))
plt.show()
```
