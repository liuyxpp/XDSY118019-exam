问题1：给定正整数n，打印出其考拉兹序列。

思路：数列递推

```python
def collatz_sequence(n): #定义Collatz Sequence
    list=[n]
    while n > 0:
        if n % 2 == 0: #判定n的奇偶性
            n = int(n/2)
            list.append(n)
        else:
            n= 3*n + 1
            list.append(n)
        if n == 1: #序列的最后一个数为1
            break
    return list
n=153673 #输入给定的正整数n
print(collatz_sequence(n)) #打印出Collatz Sequence
```

问题2：对正整数n=3,6,9,12,15,18,21,24,27,30，以它们的考拉兹序列的序号为横轴，序列值为纵轴，画出这些序列的收敛图。

思路：对上述每一个n，作出考拉兹序列的值的走势图

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np #引用需要的包
n=3
while n < 33: 
    x=np.arange(1,len(collatz_sequence(n))+1)
    y=np.array(collatz_sequence(n))
    plt.scatter(x, y,s=4) #描点
    plt.plot(x,y,linewidth=2.0) #连成折线
    plt.xlabel('n') #横坐标为序列的序号
    x_major_locator=MultipleLocator(1)
    ax=plt.gca()
    ax.xaxis.set_major_locator(x_major_locator) #横坐标刻度间距设为1
    plt.xlim(0, len(collatz_sequence(n))+1)
    plt.ylabel('C n') #纵坐标为对应的序列的值
    plt.show()
    n=n+3 #循环
    if n==27: #n=27时序列很长，单独处理一下
        x=np.arange(1,len(collatz_sequence(n))+1)
        y=np.array(collatz_sequence(n))
        plt.scatter(x, y,s=2)
        plt.plot(x,y,linewidth=1.0)
        plt.xlabel('n')
        x_major_locator=MultipleLocator(10)
        y_major_locator=MultipleLocator(500)
        ax=plt.gca()
        ax.xaxis.set_major_locator(x_major_locator)
        ax.yaxis.set_major_locator(y_major_locator)
        plt.xlim(0, len(collatz_sequence(n))+1)
        plt.ylabel('C n')
        plt.show()
        n=n+3
    if n==33: #结束循环
        break
```

问题3：对考拉兹序列进⾏有意义的分析。

考虑如下问题：对于给定的正整数m，在小于m的所有正整数里，哪一个正整数生成的考拉兹序列的长度最长。

思路：递推比较

```python
mymax=1 #定义对应的序列长度最大值，n=1时序列长为1
answer=1 #定义使得序列长度取得最大值的n,从n=1开始
n=2
m=10000 #输入给定的正整数m
while n < m :
    if mymax <= len(collatz_sequence(n)): #递推判定
        mymax=len(collatz_sequence(n))
        answer=n
        n=n+1
    else:
        n=n+1
    if n == m: #结束循环
        break
print(answer,mymax) #输出答案
```
