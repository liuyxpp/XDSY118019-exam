```python
#import需要的库
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#用于计算下一位置的函数
def fun1(x, y, a, b):
    return (1-a*x*x+y, b*x)
```

```python
#输入各项的值
a = float(input("Enter a:"))
b = float(input("Enter b:"))
list_u0 = input("Enter ux_0, uy_0:").split(",")     #ux_0和uy_0之间用英文逗号隔开
ux_0 = float(list_u0[0])
uy_0 = float(list_u0[1])
ux = [ux_0]
uy = [uy_0]
N = int(input("Enter N:"))
```

```python
#递推求u[1]~u[N]
for i in range(0, N):
    ux.append(fun1(ux[i], uy[i], a, b)[0])
    uy.append(fun1(ux[i], uy[i], a, b)[1])
```

```python
#绘图
plt.plot(ux, uy, ".")
```

```python
#采用给定好的参数
a = 1.4
b = 0.3
ux = [0.0]
uy = [0.0]
#输入N
N = int(input("Enter N:"))
print("N = %d" % (N))
for i in range(0, N):
    ux.append(fun1(ux[i], uy[i], a, b)[0])
    uy.append(fun1(ux[i], uy[i], a, b)[1])
plt.plot(ux, uy, ".")
```

以上为第二问


```python
b = 0.3
N = 10000
a_list = np.linspace(0,1.4,1000)    #让a取0~1.4之间的值，步长为1.4/1000
                                    #不让a的范围更大是因为a过大会使后续坐标数据过大而溢出，a控制在0~1.4足矣
for i in a_list:                    #遍历a_list,对每个a都画出(a, ux)的图像
    ux = [0.0]
    uy = [0.0]
    for j in range(0, N):
        ux.append(fun1(ux[j],uy[j],i,b)[0])
        uy.append(fun1(ux[j],uy[j],i,b)[1])
    plt.plot([i] * (N -5000 + 1), ux[5000:], ',k')  #由于要寻找收敛轨道，故舍去前面受初始值影响的轨道，方便寻找
```

--------------------------------------------
以上为第三问
可以看见，在 a <= 1.0 时图像为一条或多条线，呈现出周期性，我们在第四问取a = 1.0

```python
a, b = 1.0, 0.3
N = 10000
ux = [0.0]
uy = [0.0]
for i in range(0, N):
    ux.append(fun1(ux[i], uy[i], a, b)[0])
    uy.append(fun1(ux[i], uy[i], a, b)[1])
plt.plot(ux, uy,".")
#验证轨道是否收敛
print(("(x1, y1) = (%f, %f)") % (ux[500],uy[500]))
print(("(x2, y2) = (%f, %f)") % (ux[501],uy[501]))
print(("(x3, y3) = (%f, %f)") % (ux[502],uy[502]))
print(("(x4, y4) = (%f, %f)") % (ux[503],uy[503]))
print(("(x5, y5) = (%f, %f)") % (ux[504],uy[504]))
```

以上为第四问
可见当a = 1.0 时，轨道收敛至一个周期为4的轨道
