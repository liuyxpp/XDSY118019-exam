```python
import numpy as np
import matplotlib.pyplot as plt


def henon_map(a, b, u0, N):
    x, y = u0
    trajectory = [(x, y)]
    for _ in range(N):
        x_new = 1 - a * x**2 + y
        y_new = b * x
        trajectory.append((x_new, y_new))
        x, y = x_new, y_new
    return trajectory


def trajectory(a, b, u0, N):
    tra = henon_map(a, b, u0, N)
    x_vals, y_vals = zip(*tra)
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, 'o', markersize=2, label=f"a={a}, b={b}")
    plt.title("Hénon Map Trajectory")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid()
    plt.show()


def orbit(b, a_range, u0, N, skip=100):
    a_values = np.arange(*a_range)
    x_vals = []

    for a in a_values:
        tra = henon_map(a, b, u0, N)
        x_vals.append([x for x, _ in tra[skip:]])

    plt.figure(figsize=(10, 6))
    for i, a in enumerate(a_values):
        plt.plot([a] * len(x_vals[i]), x_vals[i], ',k', markersize=1)
    plt.title("Hénon Map Orbit Diagram")
    plt.xlabel("a")
    plt.ylabel("x")
    plt.grid()
    plt.show()
```

```python
a, b, u0, N = 1.4, 0.3, (0, 0), 10000
trajectory(a, b, u0, N)
```

```python
orbit(b=0.3, a_range=(1.0, 1.4, 0.001), u0=(0, 0), N=1000)
```

```python
trajectory(1.01, 0.3, (0, 0), 10000)
```
