```python
import numpy as np
import matplotlib.pyplot as plt

def henon_map(a, b, u0, N):
    """
     Hnon map 

    :
        a (float): Hnon map  a
        b (float): Hnon map  b
        u0 (tuple):  (x0, y0)
        N (int): 

    :
        list:  [u0, u1, u2, ..., uN]
    """
    trajectory = [u0]
    x, y = u0

    for _ in range(N - 1): #  N - 1 
        x_next = 1 - a * x ** 2 + y
        y_next = b * x #  Hnon map 
        trajectory.append((x_next, y_next))
        x, y = x_next, y_next #  x, y 

    return trajectory

def plot_trajectory(trajectory, title="Hnon Map Trajectory"):
    """
     Hnon map 

    :
        trajectory (list): Hnon map 
        title (str): 
    """
    x_vals = [point[0] for point in trajectory]
    y_vals = [point[1] for point in trajectory]
    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, y_vals, 'o', markersize=1)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(title)
    plt.show()

def plot_orbit_diagram(b, a_min, a_max, a_steps, u0, N_plot):
    """
     Hnon map  orbit diagram

    :
        b (float): Hnon map  b
        a_min (float): a 
        a_max (float): a 
        a_steps (int): a 
        u0 (tuple):  (x0, y0)
        N_plot (int):  a 
    """
    a_values = np.linspace(a_min, a_max, a_steps) #  a 
    x_points = []
    a_points = []

    for a in a_values:
        #  N_plot 
        trajectory = henon_map(a, b, u0, N_plot) # 
        x_vals = [point[0] for point in trajectory] #  x 
        
        x_points.extend(x_vals)
        a_points.extend([a] * N_plot)

    plt.figure(figsize=(10, 8))
    plt.plot(a_points, x_points, ',k', markersize=1)
    plt.xlabel('a')
    plt.ylabel('x')
    plt.title('Orbit Diagram of Hnon Map')
    plt.show()

# Part 1:  Hnon map 
a = 1.4
b = 0.3
u0 = (0, 0)
N = 100000 # 
trajectory = henon_map(a, b, u0, N)

# Part 2:  Hnon map 
plot_trajectory(trajectory, title="Hnon Map Trajectory for a=1.4, b=0.3")

# Part 3:  Hnon map  orbit diagram
a_min = 1.0
a_max = 1.4
a_steps = 400  # 
N_plot = 500  #  a 500
plot_orbit_diagram(b, a_min, a_max, a_steps, u0, N_plot)
```

```python
# Part 4:  a 
#  orbit diagram  a=1.04 
a_periodic = 1.04
N = 1000000 #  N 
periodic_trajectory = henon_map(a_periodic, b, u0, N)
plot_trajectory(periodic_trajectory, title="Periodic Trajectory for a=1.04, b=0.3")
```
