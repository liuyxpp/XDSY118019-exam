```python
# 1.实现Bisection算法
def bisection(f, a, b, epsilon):
    while (b - a) / 2 > epsilon:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
    return (a + b) / 2
```

```python
# 2.实现Newton算法
def newton(f, df, x0, epsilon):
    x = x0
    while abs(f(x)) > epsilon:
        x = x - f(x) / df(x)
    return x
```

```python
# 3.导入相关库
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# 3.定义目标函数及其导数
def f1(x):
    return 2*x - np.tan(x)

def df1(x):
    return 2 - (1/np.cos(x))**2

def f2(x):
    return np.exp(x+1) - x - 2

def df2(x):
    return np.exp(x+1) - 1

def f3(x):
    return x**(-2) - np.sin(x)

def df3(x):
    return -2*x**(-3) - np.cos(x)
```

```python
# 3.定义绘制目标函数图像的函数
def plot_function(func, x_range, label=None, title=None, xlabel=None, ylabel=None):
    x_vals = np.linspace(x_range[0], x_range[1], 100)
    plt.plot(x_vals, func(x_vals), label=label)
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.show()
```

```python
# 3. i. 2x=tan(x), x∈[-0.2,1.4]
plot_function(f1, [-0.2, 1.4], label='2x - tan(x)', title='Function: 2x - tan(x)', xlabel='x', ylabel='f(x)')

f1_scipy_roots = fsolve(f1, [-0.2, 1.4])
f1_bisection_root1 = bisection(f1, -0.2, 0.5, 1e-10)
f1_bisection_root2 = bisection(f1, 0.5, 1.4, 1e-10)
f1_newton_root1 = newton(f1, df1, 0.5, 1e-10)
f1_newton_root2 = newton(f1, df1, 1.0, 1e-10)

print(" Scipy:")
print("     roots:",f1_scipy_roots)
print(" Bisection:")
print("     root1:",f1_bisection_root1)
print("     root2:",f1_bisection_root2)
print(" Newton:")
print("     root1:",f1_newton_root1)
print("     root2:",f1_newton_root2)
```

```python
# 3. ii. e^(x+1)=2+x, x∈[-2,2]
plot_function(f2, [-2, 2], label='exp(x+1) - x - 2', title='Function: exp(x+1) - x - 2', xlabel='x', ylabel='f(x)')

f2_scipy_roots = fsolve(f2, 0)
f2_bisection_root = bisection(f2, -2, 2, 1e-10)
f2_newton_root = newton(f2, df2, 0, 1e-10)

print(" Scipy:")
print("     roots:",f2_scipy_roots)
print(" Bisection:")
print("     root:",f2_bisection_root)
print(" Newton:")
print("     root:",f2_newton_root)
```

```python
# 3. iii. x^(-2)=sin(x), x∈[0.5,4Π]
plot_function(f3, [0.5, 4*np.pi], label='x^(-2) - sin(x)', title='Function: x^(-2) - sin(x)', xlabel='x', ylabel='f(x)')

f3_scipy_roots = fsolve(f3, [1.0, 3.0, 6.0, 10.0, 12.0])

f3_bisection_root1 = bisection(f3, 0.5, 2, 1e-10)
f3_bisection_root2 = bisection(f3, 2, 4, 1e-10)
f3_bisection_root3 = bisection(f3, 4, 8, 1e-10)
f3_bisection_root4 = bisection(f3, 8, 10, 1e-10)
f3_bisection_root5 = bisection(f3, 10, 4*np.pi, 1e-10)

f3_newton_root1 = newton(f3, df3, 1, 1e-10)
f3_newton_root2 = newton(f3, df3, 3, 1e-10)
f3_newton_root3 = newton(f3, df3, 6, 1e-10)
f3_newton_root4 = newton(f3, df3, 10, 1e-10)
f3_newton_root5 = newton(f3, df3, 12, 1e-10)

print(" Scipy:")
print("     roots:",f3_scipy_roots)
print(" Bisection:")
print("     root1:",f3_bisection_root1)
print("     root2:",f3_bisection_root2)
print("     root3:",f3_bisection_root3)
print("     root4:",f3_bisection_root4)
print("     root5:",f3_bisection_root5)
print(" Newton:")
print("     root1:",f3_newton_root1)
print("     root2:",f3_newton_root2)
print("     root3:",f3_newton_root3)
print("     root4:",f3_newton_root4)
print("     root5:",f3_newton_root5)
```

```python
# 4.定义目标函数及其导数
def f4(x):
    return x**2 - 2

def df4(x):
    return 2*x
```

```python
# 4.定义误差分析的函数
def bisection_convergence(f, a, b, epsilon, max_iterations=100):
    errors = []
    x = (a + b) / 2
    for i in range(max_iterations):
        if f(x) == 0:
            break
        elif f(x) * f(a) < 0:
            b = x
        else:
            a = x
        x = (a + b) / 2
        errors.append(abs(x - np.sqrt(2)))
        if (b - a) / 2 < epsilon:
            break
    return errors

def newton_convergence(f, df, x0, epsilon, max_iterations=100):
    errors = []
    x = x0
    for i in range(max_iterations):
        x = x - f(x) / df(x)
        errors.append(abs(x - np.sqrt(2)))
        if abs(f(x)) < epsilon:
            break
    return errors
```

```python
# 4.误差分析
errors_bisection = bisection_convergence(f4, 1, 2, 1e-10)
errors_newton = newton_convergence(f4, df4, 1.5, 1e-10)
```

```python
# 4.绘制误差分析图表
plt.figure(figsize=(12, 6))
plt.plot(errors_bisection, label='Bisection')
plt.plot(errors_newton, label='Newton')
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Error')
plt.title('Convergence Analysis for Bisection and Newton Methods')
plt.legend()
plt.show()
```
