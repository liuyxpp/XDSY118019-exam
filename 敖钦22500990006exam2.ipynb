```python
#task 1
import numpy as np

def format_number(num, decimals=0, last_digit=[]):
    '''Format a number to a string suitable for the name of a data directory.
    Args:
        num (float): a float number to be formatted.
        decimals (Optional[int]): Number of decimal places to round to. If decimals is negative, it specifies the number of positions to the left of the decimal point. Default: 0 (round to integer).
        last_digit (Optional[List(int)]): A list of single digits that is allowed for the last digit of the output number. Default: [] (all digits are allowed).
    Returns:
        str: a string represents a float number.
    '''
    decimals = int(decimals)  # ensure it is an integer
    num_rounded = round(num, decimals)
    if decimals > 0:
        format = '{:.' + repr(decimals) + 'f}'
        numstr = format.format(num_rounded)
    else:
        num_rounded = int(num_rounded)
        numstr = '{:d}'.format(num_rounded)

    numstr = list(numstr)  # to ease manipulation
    last = int(numstr[-1])

    if last_digit:
        if last not in last_digit:
            i = np.searchsorted(np.sort(last_digit), last)
            if i == 0:
                left = last_digit[0]
                right = last_digit[1]
            elif i == len(last_digit):
                left = last_digit[-2]
                right = last_digit[-1]
            else:
                left = last_digit[i - 1]
                right = last_digit[i]
            numstr[-1] = repr(left)
            num_left = float(''.join(numstr))
            numstr[-1] = repr(right)
            num_right = float(''.join(numstr))
            if abs(num) - abs(num_left) < abs(num_right) - abs(num):
                numstr[-1] = repr(left)
            else:
                numstr[-1] = repr(right)

    return ''.join(numstr)
#  input 1输入一个
for i in range(0,25):
    if(i<=5):
        a1 = np.random.random()#小数
        print("input one number:",a1)
        solv1 = format_number(a1)
        print(solv1)
    if(i>=5 and i<=10):
        a1 = np.random.random()*-1#负数
        print("input one number:",a1)
        solv1 = format_number(a1)
        print(solv1)
    if(i>=10 and i<=15):
        a1 = np.random.random()*10**i#正数
        print("input one number:",a1)
        solv1 = format_number(a1)
        print(solv1)
    if(i>=15 and i<=20):
        a1 = -1*np.random.random()*10**i#负数2
        print("input one number:",a1)
        solv1 = format_number(a1)
        print(solv1)
    if(i>=2and i<=25):
        a1 = np.random.randint(1,i)#整数
        print("input one number:",a1)
        solv1 = format_number(a1)
        print(solv1)
#  input 2输入2个
for i in range(0,25):
    deci = np.random.randint(-100,100)
    if(i<=5):
        a1 = np.random.random()
        print("input two number:",a1,deci)
        solv1 = format_number(a1,deci)
        print(solv1)
    if(i>=5 and i<=10):
        a1 = np.random.random()*-1
        print("input two number:",a1,deci)
        solv1 = format_number(a1,deci)
        print(solv1)
    if(i>=10 and i<=15):
        a1 = np.random.random()*10**i
        print("input two number:",a1,deci)
        solv1 = format_number(a1,deci)
        print(solv1)
    if(i>=15 and i<=20):
        a1 = -1*np.random.random()*10**i
        print("input two number:",a1,deci)
        solv1 = format_number(a1,deci)
        print(solv1)
    if(i>=2and i<=25):
        a1 = np.random.randint(1,i)
        print("input two number:",a1,deci)
        solv1 = format_number(a1,deci)
        print(solv1)
#  input 3输入3个
for i in range(0,25):
    deci = np.random.randint(-100,100)
    if(i<=5):
        a1 = np.random.random()
        last = np.random.randint(1,9)
        print("input three number:",a1,deci,last)
        solv1 = format_number(a1,deci,)
        print(solv1)
    if(i>=5 and i<=10):
        a1 = np.random.random()*-1
        print("input three number:",a1,deci,last)
        solv1 = format_number(a1,deci,last)
        print(solv1)
    if(i>=10 and i<=15):
        a1 = np.random.random()*10**i
        print("input three number:",a1,deci,last)
        solv1 = format_number(a1,deci,last)
        print(solv1)
    if(i>=15 and i<=20):
        a1 = -1*np.random.random()*10**i
        print("input three number:",a1,deci,last)
        solv1 = format_number(a1,deci,last)
        print(solv1)
    if(i>=2and i<=25):
        a1 = np.random.randint(1,i)
        print("input three number:",a1,deci,last)
        solv1 = format_number(a1,deci,last)
        print(solv1)
a1 = 0
print("if the nim is empty",a1,1,0)
solv1 = format_number(a1,1,0)
print(solv1)

```
