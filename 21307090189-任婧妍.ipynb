```python
#调用函数
import numpy as np
import matplotlib.pyplot as plt
```

```python
#迭代函数
def map(a, b, x_0, y_0):
    return [1-a*x_0**2+y_0, b*x_0]
#迭代过程
def trail(a, b, u_0, N):
    u = u_0#初始坐标
    trail_list = []#用于存储坐标点
    for i in range(N):
        u = map(a, b, u[0], u[1])#每次更新坐标
        trail_list.append(u)#存储坐标点
    return trail_list
```

```python
trail_list = trail(1.4, 0.3, [0,0], 100)
trail_np = np.array(trail_list)
#绘制图像
plt.plot(trail_np.T[0], trail_np.T[1])
```
